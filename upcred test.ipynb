```python
import pandas as pd
import numpy as np
```

```python
"""READING ALL THE CSV FILES i.e. PLAYER, PLAYER DATA AND SEASON STATS FILES"""

player=pd.read_csv(r"C:\Users\HP\Downloads\Players.csv")
playerdata=pd.read_csv(r"C:\Users\HP\Downloads\player_data.csv")
stats=pd.read_csv(r"C:\Users\HP\Downloads\Seasons_Stats.csv")
```

```python
player.head()
```

```python
#dropping unnmaed column of indexing
player=player.drop(['Unnamed: 0'],axis=1)
```

```python
#sorting the player table by alphabetical order
player=player.sort_values('Player')
```

```python
#sorting the playe data table by alphabetical order
playerdata=playerdata.sort_values('name')
```

```python
#merging player data to player table (only the position column as other columns are already there)
player["position"]=playerdata['position']
```

```python
player
```

```python
#creating a updated stats table containg only few columns which are reuired
ustats=pd.DataFrame(stats,columns=['Year','Player','Age','Pos','PTS','G'])
```

```python
#sorting the updated stats table by alphabetical order
ustats=ustats.sort_values('Player')
```

```python
# creating a new table named main containg only necesarry  information required to solve the problem
main=pd.DataFrame(player,columns=['Player','position'])
```

```python
main.insert(1,'Age',ustats['Age'])
main.insert(3,'Games',ustats['G'])
main.insert(4,'Points',ustats['PTS'])
```

```python
# adding a column of average points per game by each player, this will be used by us to select our best players for the team.
main['avg game pt']=main['Points']/main['Games']
```

```python
#sorting the main table by the average points per game in descending order
main=main.sort_values('avg game pt',ascending=False)
```

```python
main
```

```python
marr=main.to_numpy()
```

```python
(marr[0][2])
```

```python
#creating a list of player at each position(i.e. PG,SG,SF,C,PF) 
#the list is already sorted by average game point 

PG=SG=SF=C=PF=[]

for i in range(len(marr)):
    if marr[i][2]== 'PG':
        PG.append((marr[i][0],marr[i][4]))
        
    if marr[i][2]== 'SG':
        SG.append((marr[i][0],marr[i][4]))
        
    if marr[i][2]== 'SF':
        SF.append((marr[i][0],marr[i][4]))
        
    if marr[i][2]== 'PF':
        PF.append((marr[i][0],marr[i][4]))
        
    if marr[i][2]== 'C':
        C.append((marr[i][0],marr[i][4]))
```

```python
PG=np.array(PG)
SG=np.array(SG)
SF=np.array(SF)
PF=np.array(PF)
C=np.array(C)
```

```python
#this function takes input the spcific point sum from te user, from this points the team will be created.
# I divide the total points by 5 and use that value to select the best player from list of each position.

def teampred(totalpoints):
    team=[]
    avpt=totalpoints/5
    f0=f1=f2=f3=f4=0
    for i in range (len(PG)):
        
        if f0==0 and float(PG[i][1])<=float(avpt):
            f0=1
            team.append(('POINT GUARD: ',PG[i][0],"  ", 'POINTS: ',PG[i][1]))
            
        if f1==0 and float(SG[i][1])<=float(avpt):
            f1=1
            team.append(('SHOOTING GUARD: ',SG[i][0],"  ", 'POINTS: ',PG[i][1]) )
            
        if f2==0 and  float(SF[i][1])<=float(avpt):
            f2=1
            team.append(('SMAL FORWARD: ',SF[i][0],"  ", 'POINTS: ',PG[i][1]))
            
        if f3==0 and float(PF[i][1])<=float(avpt):
            f3=1
            team.append(('POWER FORWARD: ',PF[i][0],"  ", 'POINTS: ',PG[i][1]))
            
        if f4==0 and float(C[i][1])<=float(avpt):
            f4=1
            team.append(('CENTER: ' ,C[i][0],"  ", 'POINTS: ',PG[i][1]))    
      
    print("TOTAL POINT ENTERED:",totalpoints,"\n" )                     
    for member in team:
        print(member)
    
```
